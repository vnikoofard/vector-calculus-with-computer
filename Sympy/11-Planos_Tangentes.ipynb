## Planos Tangentes

Na GA aprendemos como construir a queação de um plano. Sabemos fazer isso usando o seu vetor tangente e um ponto dele, ou, usando tres pontos dele.

### Plano tangente e curvas de grade

Na mesma forma que para cada ponto de curva há uma linha tangente, para cada ponto de uma superficie existe um plano tangente. Tem varias formas para chegar na equação destes planos. Talvez o caminho mais facil seja pelas **curvas de grade**. 

Para uma superficie $z = f(x,y)$, se fixarmos um dos parametros $x$ and $y$ achamos uma curva de grade da superficie. Por exemplo, para $z=\sin(x^2 + y^2)$ temos as seguintes curvas
$$
x = 1 \rightarrow z=\sin(1+y^2) \\
x = 2 \rightarrow z=\sin(4+y^2) \\
x = 3 \rightarrow z=\sin(9+y^2) \\
\\
y = 1 \rightarrow z=\sin(x^2+1) \\
y = 2 \rightarrow z=\sin(x^2+4) \\
y = 3 \rightarrow z=\sin(x^2+9) \\
$$

Graficamente podemos plotar elas na seguinte forma

```python
# Creating the data
xx = np.linspace(-5, 5, 20)
yy = np.linspace(-5, 5, 20)
xGrid, yGrid = np.meshgrid(xx, yy)
RR = np.sqrt(xGrid ** 2 + yGrid ** 2)
zz = np.sin(RR)

# Creating the plot
lines = []
line_marker = dict(color='#4ca3dd', width=2)
for i, j, k in zip(xGrid, yGrid, zz):
    lines.append(go.Scatter3d(x=i, y=j, z=k, mode='lines', line=line_marker))

line_marker = dict(color='#b12600', width=2)
for i, j, k in zip(xGrid, yGrid, zz):
    lines.append(go.Scatter3d(x=j, y=i, z=k, mode='lines', line=line_marker))

layout = go.Layout(
    title='$z=sin(x^2+y^2)$',
    showlegend=False,
)
fig = go.Figure(data=lines, layout=layout)
fig.show()
```

Da para ver que de cada ponto da superficie passm duas curvas de grade. Cada um dessa curvas tem um vetor tangente que tambem são tangentes a superficie. De fato, cada ponto de uma superficie tem infinitas vetores tangentes. Dois destes vetores são os que acabamos de achar. Portanto, tendo dois vetores tangentes a superficie, usando o produto vetorial, podemos achar o vetor normal (perpendicular) a superficie. Finalmente, usando este vetor normal e as coordenadas do ponto podemos montar o plano tangente a superficie no ponto.

### Plano tangente via vetor gradiente

Outra maneira de encontrar o plano tangente para uma superfície é usar o vetor gradiente. Lembre-
se de que o vetor gradiente de uma função $z=f(x, y)$ é definido como
$$
\vec\nabla f(x,y) = \frac{\partial f}{\partial x} \hat i + \frac{\partial f}{\partial y} \hat j
$$
Hmmm. Este é um vetor bidimensional, e para definir um plano no espaço, definitivamente precisamos de vetores com três componentes. Nós contornamos isso fazendo pequenas reorganizações.
Nossa superfície é definida por $z = f( x, y)$. Poderíamos mover tudo para o mesmo lado e dizer que nossa superfície é definida pela equação
$$
z - f(x,y) =0
$$
Você pode estar pensando que este é um passo estranho; de alguma forma, é uma maneira mais complicada de escrevê-lo. Isso é verdade, e vai ficar um pouco pior. Em vez de dizer que nossa superfície é definida por essa equação, vamos definir uma nova função :
$$
g(x,y,z) = z - f(x,y)
$$
E agora dizemos que nossa função é a superficies de níveis definido por $g ( x, y, z) = 0$. Então, agora, em vez da afirmação simples e simples $z = f ( x, y)$, estamos de repente falando sobre conjuntos de níveis. Caramba! Há uma razão para isso, no entanto ...

Observe que $g ( x, y, z)$ é uma função de três variáveis, e seu gradiente é
$$
\vec\nabla g(x,y,z) = -\frac{\partial f}{\partial x} \hat i - \frac{\partial f}{\partial y} \hat j + \hat k
$$

- Podemos começar com um ponto $( x, y, z)$ que esteja na superfície de nível $g ( x, y, z) = 0$.
- Por causa da maneira como definimos $g ( x, y, z)$, isso é o mesmo que dizer que o ponto está na superfície $z = f ( x, y)$.
- Além disso, $\vec\nabla g (x, y, z)$ será perpendicular (ou normal) à nossa superfície nesse ponto.
- Em particular, podemos usar o $\vec\nabla g (x, y, z)$ como um vetor normal para definir a equação do nosso plano tangente!

Por exemplo para a superficie $z = \sin(x^2 + y^2)$ temos
$$
g = g(x,y,z) = z - \sin(x^2 + y^2) \rightarrow \vec\nabla g = - 2x \sin(x^2 + y^2) \hat i - 2y \sin(x^2 + y^2) \hat j + \hat k
$$
Esse vetor é perpendicular a superficie em todos os pontos $(x,y)$

```python
def f(x,y):
    return sp.sin(x**2 + y**2)
```

```python
f(x,y)
```

O vetor normal a superficie é calculada

```python
vn = av.gradient(z-f(x,y), (x,y,z))
```

```python
vn
```

Os componentes do vetor são

```python
a = vn & R.i
b = vn & R.j
c = vn & R.k
```

Finalmene, a equação do plano tangente no ponto $(x_1,y_1,z_1)$ é 

```python
plano = sp.Eq(a*(x-x_1)+b*(y-y_1)+ c*(z-z_1),0)
```

```python
plano
```

E por exemplo, no ponto $(1, 2, f(2,2))$ é 

```python
a1 = a.subs(x,1).subs(y,2).subs(z,f(1,2))
b1 = b.subs(x,1).subs(y,2).subs(z,f(1,2))
c1 = c.subs(x,1).subs(y,2).subs(z,f(1,2))
```

```python
a1
```

```python
plano1 = plano.subs(a,a1).subs(b,b1).subs(c,c1).subs(x_1,1).subs(y_1, 2).subs(z_1,f(1, 2))
```

```python
sp.N(plano1,2)
```

```python
res = sp.solve(plano1, z, dict=True)
```

Isolando a variavel $z$ temos 

```python
res[0][z]
```

Ou seja, a equação do plano tangente é $z=2x\cos(5) + 4y\cos(5)-10\cos(5)+\sin(5)$

```python
fig = go.Figure()
fig = av.plot3d(f(x,y), (x,-3,3),(y,-3,3))
fig = av.plot3d(res[0][z], (x,0,2), (y,0,3),fig=fig, points=20, title=f"${res[0][z]}$")
fig.add_scatter3d(x=[1], y=[2], z=[float(f(1,2))], marker_size=5)
```
